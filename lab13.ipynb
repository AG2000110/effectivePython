```python
from sklearn.datasets import load_svmlight_file

X, y = load_svmlight_file("C:\\Users\\aleks\\Desktop\\efektywnyPython\\lab1\\breast-cancer.txt")
X = X.toarray()
```

```python
import numpy as np
#Zadanie 1.1

y = np.where(y == 2, -1, y)
y = np.where(y == 4, 1, y)

#Zadanie 1.2

X = (X - X.min(axis=0))/(X.max(axis=0) - X.min(axis=0))
```

```python
#Zadanie 2

import matplotlib.pyplot as plt

x1 = X[:, 0]
x2 = X[:, 1]

plt.figure(figsize=(8, 6))

plt.scatter(x1[y == -1], x2[y == -1], color='lightgreen', s=300, alpha=1, label='-1')
plt.scatter(x1[y == 1], x2[y == 1], color='purple', s=100, alpha=1, label='1')

plt.title('Wizualizacja danych')
plt.xlabel('Wymiar 1')
plt.ylabel('Wymiar 2')

plt.legend()

plt.show()

```

```python
#Zadanie 3

X_T = X.T

X_T_y = np.dot(X_T,y)

X_T_X = np.dot(X_T, X)

X_T_X_1 = np.linalg.inv(X_T_X)

B = np.dot(X_T_X_1, X_T_y)
```

```python
#Zadanie 4

import matplotlib.pyplot as plt

BX_pos = np.dot(B, X[y == 1].T)
BX_neg = np.dot(B, X[y == -1].T)

plt.figure(figsize=(8, 6))
plt.hist(BX_pos, bins=30, alpha=0.5, label='y=1', color='green')
plt.hist(BX_neg, bins=30, alpha=0.5, label='y=-1', color='purple')

plt.title('Histogram BX dla rnych klas')
plt.xlabel('Warto BX')
plt.ylabel('Liczba prbek')

plt.legend()

plt.show()
```

```python
#Zadanie 5

BX = np.dot(B, X.T)

y_pred = np.where(BX<0, -1, 1)
```

```python
#Zadanie 6

accuracy = np.mean(y_pred == y)

print(f'Dokadno modelu: {accuracy * 100:.2f}%')
```

```python
#Zadanie 7

X_T = X.T

X_T_y = np.dot(X_T,y)

X_T_X = np.dot(X_T, X)
C = 1
I = np.eye(X.shape[1])
C_I = 1/C * I

X_T_X_1 = np.linalg.inv(X_T_X + C_I)


B = np.dot(X_T_X_1, X_T_y)
```

```python
BX_pos = np.dot(B, X[y == 1].T)
BX_neg = np.dot(B, X[y == -1].T)

plt.figure(figsize=(8, 6))
plt.hist(BX_pos, bins=30, alpha=0.5, label='y=1', color='green')
plt.hist(BX_neg, bins=30, alpha=0.5, label='y=-1', color='purple')

plt.title('Histogram BX dla rnych klas')
plt.xlabel('Warto BX')
plt.ylabel('Liczba prbek')

plt.legend()

plt.show()
```

```python
BX = np.dot(B, X.T)

y_pred = np.where(BX<0, -1, 1)
```

```python
accuracy = np.mean(y_pred == y)

print(f'Dokadno modelu: {accuracy * 100:.2f}%')
```
