```python
if us_baby_names['Gender'].value_counts()['F'] > us_baby_names['Gender'].value_counts()['M']:
    print("Więcej jest dziewczynek.")
elif us_baby_names['Gender'].value_counts()['F'] < us_baby_names['Gender'].value_counts()['M']:
    print("Więcej jest chłopców.")
else:
    print("Jest tyle samo dziewczynek co chłopców.")
```

```python
best_name = us_baby_names['Name'].value_counts().head(10)
print(best_name)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
best_name.plot(kind='bar')
plt.title('10 najczęściej występujących imion')
plt.xlabel('Imię')
plt.ylabel('Liczba')
plt.show()
```

```python
girls = us_baby_names[us_baby_names['Gender'] == "F"]

best_name_girls = girls['Name'].value_counts().head(10)
print(best_name_girls)
```

```python
plt.figure(figsize=(12, 6))
best_name_girls.plot(kind='bar')
plt.title('10 najczęściej występujących imion dziewczynek')
plt.xlabel('Imię')
plt.ylabel('Liczba')
plt.show()
```

```python
boys = us_baby_names[us_baby_names['Gender'] == "M"]

best_name_boys = boys['Name'].value_counts().head(10)
print(best_name_boys)
```

```python
plt.figure(figsize=(12, 6))
best_name_boys.plot(kind='bar')
plt.title('10 najczęściej występujących imion chłopców')
plt.xlabel('Imię')
plt.ylabel('Liczba')
plt.show()
```

```python
unique_names = us_baby_names['Name'].nunique()
print(unique_names)
```

```python
import numpy as np

def generator(data):
    names = data['Name'].value_counts().index
    prob = data['Name'].value_counts().values
    prob = prob/ prob.sum()
    return np.random.choice(names, p=prob)
    
print(generator(us_baby_names))
```

```python
top_name = us_baby_names['Name'].value_counts().idxmax()
print(top_name)
```

```python
name_counts = us_baby_names['Name'].value_counts()

mean_name = name_counts.mean()
median_name = name_counts.median()
std_name = name_counts.std()

print(mean_name)
print(median_name)
print(std_name)
```
